```python
sample_df =df_spotify.sample(int(0.004*len(df_spotify)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df, y = "loudness", x = "energy",
            color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness",
            color = "c").set(title = "popularity vs acousticness correlation")
```

```python
plt.title("duration of the song in different genres")
sns.color_palette("rocket")
sns.barplot(y= 'track_genre', x= 'duration_ms', data=df_spotify)
plt.xlabel("duration in seconds")
plt.ylabel("genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,4))
famous = df_spotify.sort_values("popularity", ascending = False).head(15)
sns.barplot(y='track_genre', x='popularity', data = famous).set(title= "Top 5 genres by popularity")
```
